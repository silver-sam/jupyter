```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#x-accident severity
#y- factors influencing it

df = pd.read_csv('/home/silver/Desktop/jupyter/accidents.csv')
print(df.head(20))
```

```python
df = df.dropna(subset=['Accident_severity'])
```

```python
df.info()
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

selected_columns = [
    'Day_of_week',
    'Age_band_of_driver',
    'Sex_of_driver',
    'Weather_conditions',
    'Light_conditions',
    'Number_of_vehicles_involved',
    'Number_of_casualties',
    'Accident_severity'
]

df_selected = df[selected_columns].copy()

categorical_cols = [
    'Day_of_week',
    'Age_band_of_driver',
    'Sex_of_driver',
    'Weather_conditions',
    'Light_conditions',
    'Accident_severity'
]

#converting
le = LabelEncoder()
for col in categorical_cols:
    df_selected[col] = le.fit_transform(df_selected[col].astype(str))

print(df_selected.head())

```

```python
df_selected.head(5)
```

```python
X = df_selected[['Day_of_week', 'Age_band_of_driver', 'Sex_of_driver',
                'Weather_conditions', 'Light_conditions',
                'Number_of_vehicles_involved', 'Number_of_casualties']]


y = df_selected['Accident_severity']


print(X)
print(y)
```

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred1 = regressor.predict(X_test)
print(y_pred1)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred1)
#test the accuracy of the model
```

```python
pred_y_df = pd.DataFrame({'Actual Value':y_test, 'predicted value':y_pred1, 'Difference': y_test-y_pred1})
pred_y_df[0:20]
```

```python
from sklearn import metrics
import numpy as np

#Assuming y_test and y_pred are NumPy arrays or Pandas Series
mae = metrics.mean_absolute_error(y_test, y_pred1)
mse = metrics.mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)  # RMSE is the square root of MSE
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import joblib
import pickle 
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)  

joblib.dump(model, 'accidents_model.pkl')

#Alternatively, you can save the model using pickle
with open('accidents_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
