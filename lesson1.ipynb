```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('/home/silver/Documents/school.csv')print(df.head())
```

```python

df.groupby('Gender')['Test Score'].mean().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Average Test Score by Gender')
plt.ylabel('Average Score')
plt.show()
```

```python

plt.figure(figsize=(8,6))
plt.boxplot(df['Age'], vert=False)
plt.title('Box plot of age')
plt.xlabel('age')
plt.show()
```

```python
import numpy as np
import seaborn as sns

numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('correlation pliot')
plt.show()
```

```python
correlationtable = numeric_df.corr()
# Display the correlation table
print("Correlation Table for My Data:")
print(correlationtable)
```

```python
column_name = 'Age'
unique_counts = df[column_name].value_counts()
# Create a line graph to visualize the unique counts

plt.figure(figsize=(7, 7))
plt.plot(unique_counts.index, unique_counts.values, marker='o', linestyle='dotted',color='red')
plt.title(f'Unique Counts of {column_name}')
plt.xlabel(column_name)
plt.ylabel('Count')
# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
